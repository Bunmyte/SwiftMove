```python
def number_of_casualties (x):
    if x>=0 and x<=10:
        return '0'
    elif x>10 and x<=20:
        return '1'
    elif x>20 and x<=30:
        return '2'
    elif x>30 and x<=40:
        return '3'
    elif x>40 and x<=50:
        return '4'
    elif x>50 and x<=60:
        return '5'
    elif x>60 and x<=70:
        return '6'
    elif x>70 and x<=80:
        return '7'
    elif x>80 and x<=90:
        return '8'
    else:
        return '9'
accidents['Number_of_Casualties']=accidents['Number_of_Casualties'].apply(number_of_casualties)
```
