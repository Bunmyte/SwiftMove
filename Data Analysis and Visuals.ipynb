```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Accident_2012_2014=pd.read_csv('accidents_2012_to_2014.csv')
```

```python
#Accident_2012_2014['Day_of_Week']
```

```python
Accident_data=Accident_2012_2014.copy()
```

```python
#Accident_data.sortby(['Day_of_Week'],inplace=True)
```

```python
#help('sortby')
```

```python
Accident_data['Accident_Severity'].replace([1,2,3],['Low','Low','High'],inplace=True)
Accident_data['Urban_or_Rural_Area'].replace([1,2],['Urban','Rural'],inplace=True)
```

```python
Accident_data.sort_values( by = 'Day_of_Week', ascending=True, inplace=True)
```

```python
Accident_data['Day_of_Week']
```

```python
Accident_data['Day_of_Week'].replace([1,2,3,4,5,6,7],['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'],inplace=True)
```

```python
#Accident_data['Day_of_Week'].value_counts()
```

```python
#Accident_data['Number_of_Casualties'].value_counts()
```

```python
def convert_number_of_casualties (x):
    if x>=0 and x<=1:
        return 'One'
    elif x>1 and x<=10:
        return 'Between Two and Ten'
    else:
        return 'Above Ten'
Accident_data['Number_of_Casualties']=Accident_data['Number_of_Casualties'].apply(convert_number_of_casualties)
```

```python
#Accident_data['Number_of_Casualties'].value_counts()
```

```python
sns.set_style('whitegrid')
```

```python
#sns.catplot(hue='Number_of_Casualties',x='Accident_Severity',kind='count',data=Accident_data);
```

```python
#Accident severity is highest on Fridays
fig1=sns.catplot(x='Day_of_Week',hue='Number_of_Casualties',kind='count',data=Accident_data,aspect=3);
fig1.savefig('Accident_vs_Day of the week')
```

```python
#Accidents are most severe at speed limit 30
fig2=sns.catplot(x='Speed_limit',hue='Number_of_Casualties',kind='count',data=Accident_data);
fig2.savefig('Accident_vs_Speed Limit')
```

```python
#Accident are most severe in the urban areas
fig3=sns.catplot(x='Urban_or_Rural_Area',hue='Number_of_Casualties',kind='count',data=Accident_data);
fig3.savefig('Accident_vs_Urban or Rural Area')
```

```python
#Accidents are most severe where there are no pedestrian crossing human facilities within 50 metres
fig4=sns.catplot(hue='Number_of_Casualties',x='Pedestrian_Crossing-Human_Control',kind='count',data=Accident_data, aspect=3);
fig4.savefig('Accident_vs_pedestrian crossing human facilities')
```

```python
#Accidents are most severe where there are no pedestrian crossing physical facilities within 50 metres
fig5=sns.catplot(x='Pedestrian_Crossing-Physical_Facilities',hue='Number_of_Casualties',kind='count',data=Accident_data, aspect=4);
fig5.savefig('Accident_vs_pedestrian crossing physical facilities')
```

```python
#Accidents are most severe in daylight with street light possible
fig6=sns.catplot(x='Light_Conditions',hue='Number_of_Casualties',kind='count',data=Accident_data,aspect=3);
fig6.savefig('Accident_vs_Light Conditions')
```

```python
#Accidents are most severe in fine weather without high winds
fig6=sns.catplot(x='Weather_Conditions',hue='Number_of_Casualties',kind='count',data=Accident_data,aspect=4);
fig6.savefig('Accident_vs_Weather_Conditions')
```

```python
#Accidents are most severe in dry Road Conditions
fig7=sns.catplot(x='Road_Surface_Conditions',hue='Number_of_Casualties',kind='count',data=Accident_data,aspect=3);
fig7.savefig('Accident_vs_Road_Conditions')
```

```python
#Accidents are most severe where there are no special road Conditions
fig8=sns.catplot(x='Special_Conditions_at_Site',hue='Number_of_Casualties',kind='count',data=Accident_data,aspect=3);
fig8.savefig('Accident_vs_Special Road Conditions')
```

```python
#Accidents reduced in significantly in 2013 and incresed a bit in 2014
fig9=sns.catplot(x='Year',hue='Number_of_Casualties',kind='count',data=Accident_data,aspect=3);
fig9.savefig('Accident_vs_Year')
```

```python
#Accidents are most severe on Single Carriage Way
fig10=sns.catplot(x='Road_Type',hue='Number_of_Casualties',kind='count',data=Accident_data,aspect=3);
fig10.savefig('Accident_vs_Road Type')
```

```python
#Accidents are most severe when there is nothing on carriage way
fig11=sns.catplot(x='Carriageway_Hazards',hue='Number_of_Casualties',kind='count',data=Accident_data,aspect=3);
fig11.savefig('Accident_vs_Carriage Way')
```

```python
Accident=Accident_2012_2014.copy()
```

```python
def convert_number_of_casualties (x):
    if x>=0 and x<=10:
        return '0'
    elif x>10 and x<=20:
        return '1'
    elif x>20 and x<=30:
        return '2'
    elif x>30 and x<=40:
        return '3'
    elif x>40 and x<=50:
        return '4'
    elif x>50 and x<=60:
        return '5'
    elif x>60 and x<=70:
        return '6'
    elif x>70 and x<=80:
        return '7'
    elif x>80 and x<=90:
        return '8'
    else:
        return '9'
Accident['Number_of_Casualties']=Accident['Number_of_Casualties'].apply(convert_number_of_casualties)
```

```python
Accident['Pedestrian_Crossing-Human_Control'].replace(['None within 50 metres','Control by other authorised person','Control by school crossing patrol'],[0,1,2],inplace=True)
```

```python
Accident['Pedestrian_Crossing-Physical_Facilities'].replace(['No physical crossing within 50 meters','Pedestrian phase at traffic signal junction','non-junction pedestrian crossing','Zebra crossing','Central refuge','Footbridge or subway'],[0,1,2,3,4,5],inplace=True)
```

```python
Accident['Light_Conditions'].replace(['Daylight: Street light present','Darkness: Street lights present and lit','Darkeness: No street lighting','Darkness: Street lighting unknown','Darkness: Street lights present but unlit'],[0,1,2,3,4],inplace=True)
```

```python
Accident['Weather_Conditions'].replace(['Fine without high winds','Raining without high winds','Unknown','Raining with high winds','Raining with high winds','Fine with high winds','Snowing without high winds','Fog or mist','Snowing with high winds'],[0,1,2,3,4,5,6,7,8],inplace=True)
```

```python
Accident['Accident_Severity'].replace([1,2,3],[0,0,1],inplace=True)
```

```python
Accident['Road_Surface_Conditions'].replace(['Dry','Wet/Damp','Frost/Ice','Snow','Flood (Over 3cm of water)'],[0,1,2,3,4],inplace= True)
```

```python
Accident['Special_Conditions_at_Site'].replace(['None','Roadworks','Mud','Ol or diesel','Road surface defective','Auto traffic singal out','Permanent sign or marking defective or obscured','Auto traffic signal partly defective'], [0,1,2,3,4,5,6,7], inplace = True)
```

```python
Accident['Carriageway_Hazards'].replace(['None','Other object in carriageway ','Any animal (except a ridden horse)','Pedestrian in carriageway (not injured)','Involvement with previous accident','Dislodged vehicle load in carriageway'],[0,1,2,3,4,5],inplace = True)
```

```python
Accident['Road_Type'].replace(['Single carriageway','Dual carriageway','Roundabout','One way street','Slip road','Unknown'],[0,1,2,3,4,5],inplace=True)
```
