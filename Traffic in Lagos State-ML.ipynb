```python
accidents.drop(['Accident_Index','Location_Northing_OSGR', 'Location_Easting_OSGR','Longitude', 'Latitude','Date', 'Local_Authority_(Highway)','Junction_Detail', 'Junction_Control','Time', 'Carriageway_Hazards', '1st_Road_Class', '1st_Road_Number', '2nd_Road_Class',
       '2nd_Road_Number','LSOA_of_Accident_Location', 'Local_Authority_(District)'], axis = 1, inplace = True)
```

```python
accidents.info()
```

```python
accidents['Road_Surface_Conditions'].fillna(accidents['Road_Surface_Conditions'].value_counts().index[0],inplace = True)
```

```python
 accidents['Special_Conditions_at_Site'].fillna(accidents['Special_Conditions_at_Site'].value_counts().index[0],inplace = True)
```

```python
accidents['Did_Police_Officer_Attend_Scene_of_Accident'].fillna(accidents['Did_Police_Officer_Attend_Scene_of_Accident'].value_counts().index[0],inplace = True)
```

```python
accidents.info()
```

```python
mapping = {'Pedestrian_Crossing-Physical_Facilities': 'Pedestrian_Crossing_PF', 'Pedestrian_Crossing-Human_Control':'Pedestrian_Crossing_HC', 'Did_Police_Officer_Attend_Scene_of_Accident':'Police_Presence'}
accidents.rename(columns = mapping, inplace = True)
```

```python
accidents
```

```python
accidents['Number_of_Casualties'].replace({1:0, 2:0, 3:1},inplace = True)
```

```python
accidents['Number_of_Casualties'].unique()
```

```python
def number_of_casualties (x):
    if x >= 0 and x <= 6:
        return '1'
    elif x > 6 and x <=17:
        return '2'
    elif x > 17 and x <= 93:
        return '3'
    else:
        return ''
```

```python
accidents['Number_of_Casualties'] = accidents['Number_of_Casualties'].apply(number_of_casualties)
accidents['Number_of_Casualties']
```

```python
accidents['Number_of_Casualties'].value_counts()
```

```python
dataset = pd.get_dummies(accidents,columns=['Road_Type', 'Pedestrian_Crossing_HC', 'Pedestrian_Crossing_PF', 'Light_Conditions', 'Weather_Conditions', 'Road_Surface_Conditions', 'Special_Conditions_at_Site', 'Police_Presence'])
dataset
```

```python
x=dataset.drop(['Number_of_Casualties'], 1)
y=dataset['Number_of_Casualties']
```

```python
x.info()
```

## TRAINING AND PREDICTING

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=101)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

norm = MinMaxScaler().fit(x_train)

x_train_norm = norm.transform(x_train)

x_test_norm = norm.transform(x_test)
```

```python
pd.DataFrame(x_test_norm).head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
pd.DataFrame(predictions).head(5)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
cm_log = confusion_matrix(y_test, predictions)
cm_log
```

```python
accuracy_score = accuracy_score(y_test, predictions)
print(accuracy_score*100)
```

```python
print(classification_report(y_test,predictions))
```
